```python
#ploting grapgh using seaborn here by using countplot it will count the value and you have to pass 2 agrument 'coloumn_name' and 'dataframe'
plt.figure(figsize=(4,5))
custom_palette = ['#1f77b4', '#ff7f0e']
ax = sns.countplot(x ='Gender' ,data =df,palette=custom_palette ,width =0.5)


#above code will plot the graph but if you want numbers to be shown on bars then follow below method
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# above graph shows us more orders are placed by female byers than male
```

```python
#group by to check perchasing power
df.groupby(['Gender'],as_index =False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
salaes_gen=df.groupby(['Gender'],as_index =False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(4,5))
pg=sns.barplot(x='Gender',y='Amount',data =salaes_gen,palette=custom_palette,width=0.6)
```

```python
# these graph shows parchaching power (amount Spend) of female is more than male
```

```python
ax1= sns.countplot(x='Age Group' ,hue ='Gender' ,data =df)

for bar in ax1.containers:
    ax1.bar_label(bar)
```

```python
#from the above graph we can conclude that age group 26-30 are placing more order
```

```python
salaes_age=df.groupby(['Age Group'],as_index =False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data =salaes_age,palette="tab10")
```

```python
# from these graph we can conculde that people of age group having 26-35 have more purchsing power (Spending capacity)
# and in second number it is 36-45  .
```

```python
salaes_age=df.groupby(['State'],as_index =False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(22,4)})
sns.barplot(x='State',y='Orders',data =salaes_age,palette="tab10")
```

```python
# From above we can conclude that people from uttar Pradesh are placing more number of orders
```

```python
salaes_age=df.groupby(['State'],as_index =False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(22,4)})
sns.barplot(x='State',y='Amount',data =salaes_age ,palette="tab10")
```

```python
sales_state = df.groupby('Product_Category')['Amount'].sum().nlargest(10).sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(12, 7))
ax.pie(sales_state, labels=sales_state.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Add a title
plt.title('Top 10 Products by Industry')

# Show the plot
plt.show()
```
